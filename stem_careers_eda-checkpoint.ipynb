```python
csv_clean = csv[csv["basesalary"] <= 260000].copy()
```

```python
# Triangle Correlation Heatmap
sns.set_style("white")
plt.figure(figsize=(15, 10))
mask = np.triu(np.ones_like(csv_clean.corr(), dtype=bool))

heatmap = sns.heatmap(csv_clean.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16)
plt.yticks(size=12)
plt.show()
```
